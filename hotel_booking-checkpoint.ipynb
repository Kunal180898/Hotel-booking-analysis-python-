```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("hotel_bookings.csv")
df
```

```python
df.info()
```

# Data cleaning

    # removing duplicate rows

```python
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
df.shape
```

    # handling missing values

```python
df.isna().sum()
```

```python
df["children"] = df["children"].fillna(0)
df["children"].isna().sum()
```

```python
df["country"] = df["country"].fillna("Unknown")
df["country"].isna().sum()
```

```python
# replace with 0 to indicate no agent or direct booking as this column has ID of agent
df["agent"] = df["agent"].fillna(0)
df["agent"].isna().sum()
```

```python
# deleting company column which has too much missing value
del df["company"]
```

```python
df.info()
```

    # converting data-types of column

```python
# converting year
df["arrival_date_year"] = pd.to_datetime(df["arrival_date_year"],format="%Y").dt.year

# coverting month
df["arrival_date_month"] = pd.to_datetime(df["arrival_date_month"],format="%B")
df["arrival_date_month"] = df["arrival_date_month"].dt.month

# converting day (date)
df["arrival_date_day_of_month"] = pd.to_datetime(df["arrival_date_day_of_month"],format="%d").dt.day

# converting reservation_status_date
df["reservation_status_date"] = pd.to_datetime(df["reservation_status_date"],format="%Y-%m-%d")
```

    # replacing values

```python
# for is_canceled column
df["is_canceled"] = df["is_canceled"].replace({0:"Not Canceled",1:"Canceled"}).astype(str)

# for is_repeated_guest column
df["is_repeated_guest"] = df["is_repeated_guest"].replace({0:"Not Repeated",1:"Repeated"}).astype(str)

# for market_segment column
df["market_segment"] = df["market_segment"].str.replace("TA","").str.replace("TO","").str.replace("/","")

# for distribution_channel column
df["distribution_channel"] = df["distribution_channel"].str.replace("TA","Travel Agents").str.replace("TO","Tour Operators")
```

```python
df.head()
```

    # inserting new arrival_date column 

```python
df["arrival_date"] = df["arrival_date_year"].astype(str)+"-"+df["arrival_date_month"].astype(str)+"-"+df["arrival_date_day_of_month"].astype(str)
df["arrival_date"] = pd.to_datetime(df["arrival_date"],format="%Y-%m-%d")
df.insert(3, 'arrival_date', df.pop('arrival_date'))

# Note:- This new column has been added for date clean purpose, however seperate year, month and day column will be more useful for analysis
```

```python
df["arrival_date"] 
```

```python
df.info()
```

```python
# Adding season column using month number

def get_season(month):
    if month in [12, 1, 2]:
        return 'Winter'
    elif month in [3, 4, 5]:
        return 'Spring'
    elif month in [6, 7, 8]:
        return 'Summer'
    elif month in [9, 10, 11]:
        return 'Fall'
    else:
        return 'Unknown'

df["Season"] = df["arrival_date_month"].apply(get_season)
```

# 10. Agent and Company Influence

    # 10.A. Agent-wise bookings: Analyze how different agents contribute to bookings and cancellations.

```python
# Total bookings per agent
agent_bookings = df.groupby('agent')['is_canceled'].size()

# Canceled bookings per agent
agent_cancellations = df[df['is_canceled'] == "Canceled"].groupby('agent')['is_canceled'].size()

# Combine total bookings and cancellations
agent_analysis = pd.concat([agent_bookings, agent_cancellations], axis=1, keys=['Total Bookings', 'Canceled Bookings']).fillna(0)

# Calculate cancellation rate for each agent
agent_analysis['Cancellation Rate (%)'] = (agent_analysis['Canceled Bookings'] / agent_analysis['Total Bookings']) * 100

# Sorting by cancellation rate
agent_analysis = agent_analysis.sort_values(by='Cancellation Rate (%)', ascending=False)

# Display the top agents contributing to cancellations
print(agent_analysis.head())
print("Agent 0 is reffering to non register agents.")
```

```python
top_20 = agent_analysis.head(20)
top_20 = top_20.sort_values(by='Total Bookings', ascending=False)
top_20 = top_20.head(20)
```

```python
# Plotting total bookings vs canceled bookings per agent
ax = top_20[['Total Bookings', 'Canceled Bookings']].plot(kind='bar', figsize=(12, 6), stacked=True)
plt.title('Agent-wise Total Bookings and Cancellations')
plt.xlabel('Agent')
plt.ylabel('Number of Bookings')
ax.set_xticklabels(top_20.index, rotation=90, ha='right', fontsize=8)
plt.show()
```

    Conclusion:- Agent 170 brings highest bookings.
